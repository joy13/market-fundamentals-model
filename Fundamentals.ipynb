```python
def isNumber(s):
    try:
        float(s)
        return True
    except ValueError:
        return False
    
def writeToFile(fname,orderkeys,corr):
    with open(fname,'wb') as csvfile:
        w=csv.writer(csvfile,delimiter=',')
        for i in range(len(orderkeys)):
            w.writerow((orderkeys[i],corr[i]))

def plotGraph(orderkeys,corr,year):
    feature_list=np.arange(0,len(orderkeys))
    plt.plot(feature_list,corr)
    plt.xlabel('Features')
    plt.ylabel('Correlations')
    plt.title(year)
    name=str(year)+'corr.png'
    plt.savefig(name)
    plt.show()
    
def getstats(Y, YTest):
    Y=np.reshape(Y,(len(Y),1))
    YTest=np.reshape(YTest,(len(YTest),1))
    diff = Y - YTest
    diff = diff**2
    summ = diff.sum()
    corr = findCorrelation(Y,YTest)
    return corr

def findCorrelation(attribute,changes):
    return np.corrcoef(attribute,changes,rowvar=0)[0,1]
```

```python
def addEvidence(dataX, dataY):
    if not dataY == None:
        data = np.zeros([dataX.shape[0],dataX.shape[1]+1])
        data[:,0:dataX.shape[1]]=dataX
        data[:,(dataX.shape[1])]=dataY
    else:
        data = dataX
    return data
    
def query(data,test):
    dataX = data[:,:(len(data[0])-1)]
    b = np.ones(len(data))
    tempX = np.column_stack([dataX,b])
    m = np.linalg.lstsq(tempX, data[:,(dataX.shape[1])])[0]
    a = np.ones(len(test))
    temp = np.column_stack([test,a])
    Yest = np.dot(temp, m)
    return Yest

def queryRidge(data,test):
    x = data[:,:-1]
    y = data[:,-1]
    clf = Ridge(alpha=1.0)
    yest = clf.fit(x, y).predict(test)
    return yest

def queryLasso(data, test):
    x = data[:,:-1]
    y = data[:,-1]
    clf = Lasso(alpha=0.1)
    yest = clf.fit(x,y).predict(test)
    return yest
```

```python
def callLearner(xtrain, ytrain, xtest,ltype):
    data = addEvidence(xtrain, ytrain)
    if ltype == 'linear':
        ypred = query(data,xtest)
    elif ltype == 'ridge':
        ypred = queryRidge(data,xtest)
    elif ltype == 'lasso':
        ypred = queryLasso(data,xtest)
    return ypred
```

```python
def getPercentageChange(sp500list,year,month,date):
    dt_start = dt.datetime(int(year), int(month), date)
    dt_end = dt.datetime(int(year)+1, int(month), date)
    ldt_timestamps = du.getNYSEdays(dt_start, dt_end, dt.timedelta(hours=16))
    dataobj = da.DataAccess('Yahoo')
    ls_symbols = dataobj.get_symbols_from_list(sp500list)
    ls_keys = ['close']
    ldf_data = dataobj.get_data(ldt_timestamps, ls_symbols, ls_keys)
    d_data = dict(zip(ls_keys, ldf_data))

    for s_key in ls_keys:
        d_data[s_key] = d_data[s_key].fillna(method='ffill')
        d_data[s_key] = d_data[s_key].fillna(method='bfill')
        d_data[s_key] = d_data[s_key].fillna(1.0)
    
    naPrice = d_data['close'].values
    percentChange=naPrice[len(naPrice)-1,:]/naPrice[0,:]-1 #Percentage change 
    pChange=dict()
    print 'total symbols before: ', len(ls_symbols)
    for i,sym in enumerate(ls_symbols):
        if percentChange[i] == 0:
            ls_symbols.remove(sym)
            continue
        else:
            pChange[sym]=percentChange[i]
    print 'total symbols after removing : ', len(ls_symbols)
    print 'symbols in dict ', len(pChange)
    return pChange
```

```python
def getFeatures_mean(fname,key_features):
    equityDict = defaultdict(list)
    featureValDict = defaultdict(list)
    rowWithFeatureNames = 0
    ls_equity = [] 
    count = 0
    data = np.zeros((500,len(key_features)))
    reader=csv.reader(open(fname,'rb'))
    for i,equity in enumerate(reader):
        if i == rowWithFeatureNames:
            available_features = equity
        else:
            index = 0
            for fid, value in enumerate(equity):
                if available_features[fid] in key_features:
                    fval = equity[fid]
                    if fval and isNumber(fval):
                        data[i,index] = fval
                    index += 1
    
        count = i
    data = data[1:count+1,:]
    means = np.mean(data,axis=0)
        
    reader=csv.reader(open(fname,'rb'))
    for i,equity in enumerate(reader):
        symbol = equity[9]
        if i == rowWithFeatureNames:
            available_features = equity
        else:
            ls_equity.append(symbol)
            index = 0
            for fid, value in enumerate(equity):
                if available_features[fid] in key_features:  
                    fval = equity[fid]
                    if fval and isNumber(fval):
                        equityDict[symbol].append(fval)
                        featureValDict[available_features[fid]].append(fval)
                    else:
                        equityDict[symbol].append(means[index])
                        featureValDict[available_features[fid]].append(means[index])
                    index += 1
    
    return equityDict, featureValDict, ls_equity
```

```python
def getPreviousFilenames(m,y):
    if m > 3:
        prev1 = str(m-3)+ str(y)
    else:
        prev1 = '12'+ str(y-1)
    if m > 6:
        prev2 = str(m-6)+ str(y)
    else:
        prev2 = str(m+6)+ str(y-1)
    return prev1, prev2
```

```python
def getSP500data(m,y):
    if m > 6:
        month = m-8
        year = y+1
        f = str(month)+ str(year)        
    else:
        month = m+4
        year = y
        f = str(month)+ str(year)
    return f, month, year
```
